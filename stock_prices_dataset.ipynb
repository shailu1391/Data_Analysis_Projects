### Deep diving into analysis

### What is the max Close price for each bank's stock throughout the time period?

```python
df.xs(key='Close',axis=1,level='Attributes').max()
```

### Creating a column for returns using the pct_change() method on the Close column

```python
returns = pd.DataFrame()
```

```python
for i in tickersList:
    returns[i+' Return'] = df['Close'][i].pct_change()
returns.head()
```

### using seaborn for pairplot

```python
import seaborn as sns
```

```python
sns.pairplot(returns[1:])
```

```python
import plotly.express as px
```

```python
px.scatter_matrix(returns)
```

### On what dates each bank stock had the best and worst single day returns

```python
## Worst returns day
returns.idxmin()
```

```python
## Best returns day
returns.idxmax()
```

### Notice if there is any significance behind the dates of rise/fall

### Which stock would you classify as the riskiest over the entire time period?
Ans : Stock with highest standard deviation for the returns. In this case Citi Group

```python
returns.std()
```

###  Which is the riskiest stock for the year 2019? 

```python
returns.loc['2019-01-01':'2019-12-31'].std() ## Citigroup
```
